```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv("D:\\GitHub\\Titanic-Machine-Learning-from-Disaster\\data\\raw_data\\train.csv")
data.head(15)
```

```python
data.isnull().sum()
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
data['Survived'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Survived')
ax[0].set_ylabel('')
sns.countplot('Survived',data=data,ax=ax[1])
ax[1].set_title('Survived')
plt.show()
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
data[['Sex','Survived']].groupby(['Sex']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs Sex')
sns.countplot('Sex',hue='Survived',data=data,ax=ax[1])
ax[1].set_title('Sex:Survived vs Dead')
plt.show()
data.groupby(['Sex','Survived'])['Survived'].count()
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
data['Pclass'].value_counts().plot.bar(color=['#CD7F32','#FFDF00','#D3D3D3'],ax=ax[0])
ax[0].set_title('Number Of Passengers By Pclass')
ax[0].set_ylabel('Count')
sns.countplot('Pclass',hue='Survived',data=data,ax=ax[1])
ax[1].set_title('Pclass:Survived vs Dead')
plt.show()
```

```python
f,ax=plt.subplots(1,3,figsize=(20,10))

data.Age.plot.hist(ax=ax[0],color='green',bins=20,edgecolor='black')
ax[0].set_title('Age overall')
x2=list(range(0,85,5))
ax[0].set_xticks(x2)
ax[0].set_ylim(0, 100)

data[data['Survived']==0].Age.plot.hist(ax=ax[1],bins=20,edgecolor='black',color='red')
ax[1].set_title('Survived= 0')
x1=list(range(0,85,5))
ax[1].set_xticks(x1)
ax[1].set_ylim(0, 100)

data[data['Survived']==1].Age.plot.hist(ax=ax[2],color='green',bins=20,edgecolor='black')
ax[2].set_title('Survived= 1')
x2=list(range(0,85,5))
ax[2].set_xticks(x2)
ax[2].set_ylim(0, 100)

plt.show()
```

```python
f,ax=plt.subplots(2,2,figsize=(20,15))
sns.countplot('Embarked',data=data,ax=ax[0,0])
ax[0,0].set_title('No. Of Passengers Boarded')
sns.countplot('Embarked',hue='Sex',data=data,ax=ax[0,1])
ax[0,1].set_title('Male-Female Split for Embarked')
sns.countplot('Embarked',hue='Survived',data=data,ax=ax[1,0])
ax[1,0].set_title('Embarked vs Survived')
sns.countplot('Embarked',hue='Pclass',data=data,ax=ax[1,1])
ax[1,1].set_title('Embarked vs Pclass')
plt.subplots_adjust(wspace=0.2,hspace=0.5)
plt.show()
```

```python
f,ax=plt.subplots(1,2,figsize=(20,8))

sns.countplot('Parch',data=data,ax=ax[0])
ax[0].set_title('Parch overall')

sns.barplot('Parch','Survived',data=data,ax=ax[1])
ax[1].set_title('Parch vs Survived')

plt.close(2)
plt.show()
```

```python
f,ax=plt.subplots(1,2,figsize=(20,8))

sns.countplot('SibSp',data=data,ax=ax[0])
ax[0].set_title('SibSp overall')

sns.barplot('SibSp','Survived',data=data,ax=ax[1])
ax[1].set_title('SibSp vs Survived')

plt.close(2)
plt.show()
```

```python
sns.heatmap(data.corr(),annot=True,cmap='RdYlGn',linewidths=0.2) #data.corr()-->correlation matrix
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.show()
```
